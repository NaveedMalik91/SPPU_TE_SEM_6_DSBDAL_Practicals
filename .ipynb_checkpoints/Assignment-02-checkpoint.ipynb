```python
#Checking the outliers with the help of boxplot (Showed as dots...)
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(df["Math_score"], ax=axes[0])

sns.boxplot(df["Reading_score"], ax=axes[1])

sns.boxplot(df["Writing_score"], ax=axes[2])


plt.tight_layout()
plt.show()
```

```python
# Print Skewness Before Handling Outliers ---
print("Skewness Before Handling Outliers:")
print("Math Score:", df['Math_score'].skew())
print("Reading Score:", df['Reading_score'].skew())
print("Writing Score:", df['Writing_score'].skew())
```

```python
#Handle or remove Outliers from the data
def handle_outliers(data, col):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace values below the lower bound with the lower bound
    data[col] = np.where(data[col] < lower_bound, lower_bound, data[col])
    
    # Replace values above the upper bound with the upper bound
    data[col] = np.where(data[col] > upper_bound, upper_bound, data[col])

# Apply the function to the 'math score' column
handle_outliers(df, 'Math_score')
handle_outliers(df, 'Reading_score')
handle_outliers(df, 'Writing_score')
df_cleaned = df.copy()
```

```python
#Checking the outliers removed or not with the help of boxplot
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(df["Math_score"], ax=axes[0])

sns.boxplot(df["Reading_score"], ax=axes[1])

sns.boxplot(df["Writing_score"], ax=axes[2])


plt.tight_layout()
plt.show()
```

```python
# Print Skewness AFter Handling Outliers ---
print("Skewness After Handling Outliers:")
print("Math Score:", df_cleaned['Math_score'].skew())
print("Reading Score:", df_cleaned['Reading_score'].skew())
print("Writing Score:", df_cleaned['Writing_score'].skew())
```

```python
#Apply log Transformation on the column: Math_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Visualize the original data for 'math score' column
sns.histplot(df['Math_score'], kde=True)
plt.title("Before Transformation")
plt.show()

# Apply log1p transformation to the 'math score' column
df['Math_score_Log'] = np.log1p(df['Math_score'])

# Visualize the transformed data for 'Math_score_Log' column
sns.histplot(df['Math_score_Log'], kde=True)
plt.title("After Transformation")
plt.show()

# Print the original 'math score' and transformed 'Math_score_Log' columns
print("\nTransformed Data:")
print(df[['Math_score', 'Math_score_Log']].head())
```

```python
#Apply minmax scaling on Reading_score column
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Visualize the original 'math score' data
sns.histplot(df['Reading_score'], kde=True)
plt.title("Before Min-Max Scaling")
plt.show()

# Apply Min-Max scaling using sklearn's MinMaxScaler
scaler = MinMaxScaler()
df['Reading_score_MinMax'] = scaler.fit_transform(df[['Reading_score']])

# Visualize the Min-Max scaled data
sns.histplot(df['Reading_score_MinMax'], kde=True)
plt.title("After Min-Max Scaling")
plt.show()

# Print the original 'math score' and Min-Max scaled 'Math_score_MinMax' columns
print("\nTransformed Data (Min-Max Scaling):")
print(df[['Reading_score', 'Reading_score_MinMax']].head())
```
