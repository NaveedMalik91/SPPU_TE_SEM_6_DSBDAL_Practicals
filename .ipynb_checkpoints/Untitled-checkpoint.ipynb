```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import datasets
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.tail(20)
```

```python
df.drop("deck",axis = 1,inplace =True)
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["embarked"])
df.reset_index(inplace = True, drop = True)
```

```python
df.isnull().sum()
```

```python
sns.histplot(data = df, x = "age")
plt.title("Distribution of ages")
plt.show()
```

```python
sns.countplot(data = df, x = "survived")
plt.title("Distribution of survival")
plt.show()
```

```python
sns.countplot(data = df, x = "sex",hue = "survived")
plt.title("Distribution of people according to gender and survival")
plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize = (15,5))
sns.boxplot(df["age"],ax = axes[0])
sns.boxplot(df["fare"],ax = axes[1])
plt.show()
```

```python
sns.boxplot(data=df, x = "sex",y="age",hue="survived",palette = "Greens")
plt.show()
```

```python
sns.countplot(data=df,x = "pclass",hue = "survived")
plt.title("Count of individuals according to passenger class survived/died")
plt.show()
```

```python
sns.countplot(data=df,x = "embarked")
plt.title("Count of individuals according Embarke station")
plt.show()
```

```python
sns.jointplot(data = df,x="fare",y="age",kind = "scatter")
plt.show()
```

```python
sns.barplot(data=df,y="age",x = "who")
plt.show()
```

```python
sns.barplot(data=df,y="age",x = "who",hue="survived")
plt.show()
```

```python
sns.lineplot(data=df,x="age",y="survived")
plt.show()
```

```python
sns.displot(data=df,x = "age",hue="class",kde=True)
plt.show()
```
