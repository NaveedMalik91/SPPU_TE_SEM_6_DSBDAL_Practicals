###### Fill the stroke, horsepower, peak-rpm and horsepower-binned with their mean values or with last valid non-null value using ffill() method

```python
#Replacing the missing values of horsepower column with mean of the column
avg_horsepower = df["horsepower"].astype("float").mean(axis = 0) #axis = 0 represents th column
print("The average horse power = ",avg_horsepower)
df["horsepower"] = df["horsepower"].fillna(avg_horsepower)
```

```python
#Replacing the missing values of peak-rpm column with mean of the column
avg_peakRpm = df["peak-rpm"] = df["peak-rpm"].astype("float").mean(axis = 0)
print("The average Peak-Rpm = ",avg_peakRpm)
df["peak-rpm"] = df["peak-rpm"].fillna(avg_peakRpm)
```

```python
#Replacing the missing values of stroke column with mean of the column
avg_stroke = df["stroke"].astype("float64").mean(axis = 0)
print("Average stroke = ",avg_stroke)
df["stroke"]= df["stroke"].fillna(avg_stroke)
```

```python
df.isnull().sum()
```

```python
#Simply drop the horsepower-binned rows
df.dropna(subset = ['horsepower-binned'],axis = 0,inplace = True)
#Reset the index after removing the NAN rows
df.reset_index(drop=True,inplace=True)
```

```python
df.isnull().sum()
```

```python
#Before scaling the column values
df[["height","width","length"]].head()
```

```python
#Scale the columns like length, width, height of the dataset using MinMax scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[["length","width","height"]]  = scaler.fit_transform(df[["length","width","height"]])
#After Scaling the column values between 0 to 1 using minmax scaler
df[["height","width","length"]].head()
```

```python
print(df["make"].unique())  #array of Unique make of cars
```

```python
#Label Encoding for the categorical values e.g. make, aspiration, body style etc
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder()
#Encode the make column with interger values like 0 1 2 3
df["make"] = label_encoder.fit_transform(df["make"])
```

```python
#Perform binning on columns to easily understand the distribution of the data of various columns
import pandas as pd
import matplotlib.pyplot as plt

# Define bins for the horsepower
# The bins here will be: 
# - Low: horsepower values between 0 and 100
# - Medium: horsepower values between 100 and 200
# - High: horsepower values above 200
bins = [0, 100, 200, float('inf')]  # Bins for Low, Medium, High
labels = ['Low', 'Medium', 'High']  # Labels for each bin

# Bin the 'horsepower' column into the defined categories
df['horsepower-binned'] = pd.cut(df['horsepower'], bins=bins, labels=labels, right=False)

```

```python
# Get the counts of each bin
horsepower_binned_counts = df['horsepower-binned'].value_counts()

# Plotting the bar chart
plt.bar(horsepower_binned_counts.index, horsepower_binned_counts.values)

# Adding labels and title
plt.xlabel("Horsepower Bins")
plt.ylabel("Count")
plt.title("Horsepower Binned Data")

# Show the plot
plt.show()
```

```python
#Perform binning on the Price column
import pandas as pd
import matplotlib.pyplot as plt
#Define bins
bins = [0,10000,20000,30000,float('inf')]
labels = ["Low","medium","High","Very High"]
df["Price-binned"] = pd.cut(df["price"],bins = bins,labels = labels,right  = False)
```

```python
#Get he counts of each Bin
price_bins_count = df["Price-binned"].value_counts()
plt.bar(price_bins_count.index, price_bins_count.values)
plt.xlabel("Price Bins")
plt.ylabel("Count")
plt.title("Price binned data")
plt.show()
```
