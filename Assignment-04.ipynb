```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.isnull().sum()
```

```python
#Preprocessing the data i.e. handle the missing values
#Fill Age column missing values with the mean value
mean_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(mean_age)

#Fill the cabin column missing values with the mode
mode_cabin = df["Cabin"].mode()[0]
df["Cabin"] = df["Cabin"].fillna(mode_cabin)

#drop left over missing values of any
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
# Visualizing the distribution of ages
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Ages')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Fare'], kde=True, bins=30, color='blue')
plt.title('Histogram for Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency (Passengers)')
plt.show()
```

```python
#PLot for the survival of passengers in terns of class of passenger on the ship like lower-3 upper-1 middle-2
sns.countplot(data=df,x='Pclass',hue='Survived')
```

```python
#Plot the graph in terms of passengers embarked and check their corresponding survival
#Embarked: where the traveler mounted from. S: Southampton, C: Cherbourg, and Q: Queenstown.
sns.countplot(data=df,x='Embarked',hue='Survived')
```

```python
#Boxplot of survived vs age showing outliers
sns.boxplot(data=df, x="Survived",showfliers=False, y="Age")
```
