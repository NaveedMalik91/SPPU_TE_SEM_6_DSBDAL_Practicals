```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Box plot for distribution of age with respect to each gender along with the information about whether they survived or not
sns.boxplot(x='sex', y='age', data=df, hue="survived")
```

#### Among male passengers younger people survived while on the other hand older people survived in the female population.

```python
#Box plot for distribution of age with respect to each gender
sns.boxplot(x='sex', y='age', data=df)
```

#### Median age of men is just below 30. Range of age is from 2 to 80
#### Median age of women is around 26. Range of age is from 2 to 65

```python
sns.jointplot(y='fare',x='survived',data=df,kind="scatter")
```

####                      Most of the expensive ticket holders survived. As 0-> died 1->survived

```python
import seaborn as sns
sns.barplot(x='sex', y='age', data=df)
plt.title('Average Age by Gender')
```

#### The barplot() is used to display the mean value for each value in a categorical column, against a numeric column. Mean Age for men is around 30 Mean Age for women is around 25.The average is reoresented by the vertical strip in the middle corr to age of lower tip


```python
#Joint Plot to see relationship between age and fare
sns.jointplot(x='age', y='fare', data=df, kind='hex')
```

#### Most of the passengers are between age 20 and 30 and most of them paid between 10-50 for the tickets.


```python
#Line plot of age vs survived
sns.lineplot(data = df,x = "age",y = "survived")
plt.show()
```
