```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
#Features with their data types
df.dtypes
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#Unique features in the dataset
df.species.unique()
```

```python
#Plotting the barplot for the species column
sns.countplot(x='species', data=df,palette='Set1',hue = "species")
plt.show()
```

#### This shows each species type is of same number i.e. 50

```python
#Histogram to illustrate the feature distribution
sns.histplot(x=df["sepal_length"],bins=7)
plt.title("Distribution of sepal_length feature in the dataset")
plt.show()
```

```python
#Histogram to illustrate the feature distribution of sepal width
sns.histplot(df["sepal_width"],bins=10)
plt.title("Histogram and distplot of sepal_width feature in the dataset")
sns.displot(data=df,hue="species",x="sepal_width",kde=True,bins=10)
plt.show()

```

```python
##Boxplots of categorical variables with other numerical variables
plt.figure(figsize=(15,15))
plt.subplot(221)
sns.boxplot(x="species", y='sepal_length', data=df,hue = "species",palette="Set2")
plt.subplot(222)
sns.boxplot(x="species", y='petal_length', data=df,hue = "species",palette="Set2")
plt.subplot(223)
sns.boxplot(x="species", y='sepal_width', data=df,hue = "species",palette="Set2")
plt.subplot(224)
sns.boxplot(x="species", y='petal_width', data=df,hue = "species",palette="Set2")
plt.show()
```

#### Species Setosa has smallest features and easily distinguishable from other species
#### Species Versicolor has average features
#### Species Virginica has largest features
#### Versicolor and Virginica are closely distributed so it easy hard to distinguish between them

```python
#Outliers for each numerical variable in the dataset
plt.figure(figsize=(12,12))
plt.subplot(221)
sns.boxplot(x='sepal_length', data=df)
plt.title("Sepal Length")
plt.subplot(222)
sns.boxplot(x='sepal_width', data=df)
plt.title("Sepal Width")
plt.subplot(223)
sns.boxplot(df["petal_length"])
plt.title("Petal Length")
plt.subplot(224)
sns.boxplot(x='petal_width', data=df)
plt.title("Petal Width")
plt.show()
```
